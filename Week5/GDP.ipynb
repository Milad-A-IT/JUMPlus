```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = (r"C:\Users\Milad\Desktop\JUMPlus\Project\gdp_csv.csv") 
```

```python
GDP = pd.read_csv(path)
```

```python
clean_GDP = GDP.dropna(subset = ["Country Name","Country Code","Year","Value"])
```

```python
north_america = ['United States','Canada','Mexico','Cuba','Panama']
```

```python
north_america_gdp = GDP[GDP['Country Name'].isin(north_america)]
print(len(north_america_gdp))
north_america_gdp['Country Code'].unique()
```

```python
plt.rcParams['figure.figsize'] = (12,12)
sns.set()
```

```python

fig, ax = plt.subplots()
fig.savefig('North_america.png', dpi=250)
gdp = north_america_gdp
for country in north_america:
    data = gdp[gdp['Country Name'] == country]
    sns.lineplot(x='Year', y='Value', data=data, ax=ax)
ax.legend(north_america, loc="right")
ax.set_title('GDP of north america countries from {:.0f} to {:.0f}'.format(gdp.Year.min(), gdp.Year.max()))
ax.legend(north_america, loc="right")
plt.show()
```
